### Performance de acordo com atributos da partida

```python
# Piso

conditions = [
    (df['surface'] == 'Hard'),
    (df['surface'] == 'Clay'),
    (df['surface'] == 'Grass')]

choices = [df['hard'], df['clay'], df['grass']]

df['surface_performance'] = np.select(conditions, choices)

df = df.drop(columns=['hard', 'clay', 'grass'])
```

```python
# Round

conditions = [
    (df['round'] == 'R128'),
    (df['round'] == 'R64'),
    (df['round'] == 'R32'),
    (df['round'] == 'R16'),
    (df['round'] == 'QF'),
    (df['round'] == 'SF'),
    (df['round'] == 'F'),
    (df['round'] == 'BR')]

choices = [df['round_of_128'], df['round_of_64'], df['round_of_32'], df['round_of_16'],
          df['quarter_final'], df['semi_final'], df['final'], df['for_bronze_medal']]

df['round_performance'] = np.select(conditions, choices)

df = df.drop(columns=['round_of_128', 'round_of_64', 'round_of_32', 'round_of_16', 'quarter_final', 'semi_final',
                    'final', 'for_bronze_medal'])
```

```python
from datetime import date


def int2date(argdate: int) -> date:
   
    year = int(argdate / 10000)
    month = int((argdate % 10000) / 100)
    day = int(argdate % 100)

    return date(year, month, day)b
```

```python
# Ajuste Data

df_match['date'] = df_match['tourney_date'].apply(lambda x: int2date(x)).apply(pd.to_datetime)
df_match.sort_values(by=['date'], inplace=True, ascending = True)
```

```python
# Ajuste Idade

df_match['winner_age'].replace(',', '.', regex=True, inplace=True)
df_match['loser_age'].replace(',', '.', regex=True, inplace=True)

df_match['winner_age'] = df_match['winner_age'].astype(float)
df_match['loser_age'] = df_match['loser_age'].astype(float)
```
